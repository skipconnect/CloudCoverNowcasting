```python
# Making a copy of the opened satellite image
geo_target = output_area.copy()
geo_BCE = output_area.copy()
geo_L1 = output_area.copy()
geo_L1_BCE = output_area.copy()

# replacing the data in the satellite image with the output tensor
geo_target.data = target[3].squeeze()[5]
geo_BCE.data = outBCE[3].squeeze()[5]
geo_L1.data = outL1[3].squeeze()[5]
geo_L1_BCE.data = outL1_BCE[3].squeeze()[5]

# Making a list wit all objects
frames = [(geo_target,"Ground Truth"), (geo_BCE,"BCE"), (geo_L1,r"L$_1$"), (geo_L1_BCE,r"L$_1+$BCE")]
```

# Generating plot

```python
data_proj = ccrs.Geostationary(central_longitude=0, 
    satellite_height=35785863.00, sweep_axis="y", globe=ccrs.Globe())
map_proj = data_proj

plt.rc('font', family='DejaVu serif')
plt.rc('font', size = 12)

# Generating a modified colormap with transparency
from matplotlib.colors import LinearSegmentedColormap

# get colormap
ncolors = 600
color_array = plt.get_cmap('gray_r')(range(ncolors))

# change alpha values
#color_array[:,-1] = np.linspace(0.0,1.0,ncolors)
thr = 0
color_array[0:thr,-1] = 0
color_array[thr:,-1] = np.linspace(0.1,0.6,(ncolors-thr))
# create a colormap object
map_object = LinearSegmentedColormap.from_list(name='cloud_aplha',colors=color_array)

# register this new colormap with matplotlib
plt.register_cmap(cmap=map_object)




fig, axis = plt.subplots(2,2, figsize=(12,10), dpi=100, subplot_kw=dict(projection=map_proj))


for i,ax in enumerate(axis.flatten()):
    #ax.set_extent([7.76, 12.2, 47.45, 54.5], crs=ccrs.PlateCarree()) # (ind fra venstre, ind fra hjre, hjde nedefra, hjde oppefra)
    ax.set_extent([7.83, 12.15, 47.45, 54.5], crs=ccrs.PlateCarree()) # (ind fra venstre, ind fra hjre, hjde nedefra, hjde oppefra)
    #ax.stock_img()
    im = frames[i][0].plot(transform = data_proj, cmap='cloud_aplha', add_colorbar=False, ax = ax, vmin=0, vmax = 1)
    ax.set_title(frames[i][1])
    ax.add_feature(cf.COASTLINE, lw=2)
    ax.add_feature(cf.BORDERS)
    gl = ax.gridlines(crs=ccrs.PlateCarree(), draw_labels=False,
                  linewidth=2, color='gray', alpha=0.5, linestyle='--')

fig.subplots_adjust(right=0.8)
cbar_ax = fig.add_axes([0.85, 0.15, 0.02, 0.7]) #Adjustment from right, adjustment from top, width of CB, height of CB
fig.colorbar(im, cax=cbar_ax)
plt.suptitle("Plotting 6th frame for random test-datapoint for different loss-functions", y=0.95, fontsize = 14) 
plt.savefig("LossPlot.pdf")
```
